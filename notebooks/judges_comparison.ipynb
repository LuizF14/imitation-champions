# Comparação entre Qwen 2.5 com e sem fine tuning

```python
import pandas as pd

qwen2_5_results = pd.read_csv("results/judge_evaluation_qwen2_5.csv")
qwen2_5_fine_tuned_results = pd.read_csv("results/judge_evaluation_qwen2_5_fine_tuned.csv")

qwen2_5_accuracy = qwen2_5_results["is_correct"].mean()
qwen2_5_fine_tuned_accuracy = qwen2_5_fine_tuned_results["is_correct"].mean()
print(f"Qwen 2.5 sem fine tuning: {qwen2_5_accuracy}")
print(f"Qwen 2.5 com fine tuning: {qwen2_5_fine_tuned_accuracy}")
```
